```python
import os
import autolingua
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
sns.set_palette("tab10")
```

```python
table_path = os.path.join(autolingua.__path__[0], '..', 'data', 'agg_analysis.csv')
df = pd.read_csv(table_path, sep=';', index_col=0)
```

```python
df
```

```python
# Select the rows with df['Estudio'] that contain 'Centro de interés'
df_ci = df[df['Estudio'].str.contains(f' - NivelEducativo')]

# Cast the index to int and set it as a new column 'legend'
df_ci['legend'] = df_ci.index.astype(int)

# Set a new column 'x' with the value between 'Centro de interés ' and ' - NivelEducativo' of column 'Estudio'
df_ci['x'] = df_ci['Estudio'].apply(lambda x: int(x.split('Centro de interés ')[1].split(' - NivelEducativo')[0]))
```

```python
# Check everything is ok
df_ci
```

```python
# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_ci, kind="bar",
    x="x", y="Longitud listas", hue="legend", alpha=1, height=5, aspect=2.5
)
g.despine(left=True)
g.set_axis_labels('Centro de interés', 'Longitud de las listas')
g.legend.set_title('Nivel educativo')

# Increase the gap between the legend and the plot
plt.subplots_adjust(right=0.89)
```
